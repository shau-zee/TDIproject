```python
price_sorted_df = df1.loc[df_less_than60]
```

```python
price_sorted_df
```

```python
price_n_exp_ratio_sorted = price_sorted_df.loc[df_exp_rat_half]
```

```python
price_n_exp_ratio_sorted.head() #df with both conditions applied
```
